```python


base_url = "https://www.shl.com/solutions/products/product-catalog/?start="

# Initialize a list to store data for all tests across all pages
all_tests_data = []

# Loop through all pages, assuming you need to fetch pages starting from 0, 12, 24, etc.
for start_value in range(0, 31 * 12, 12):  # Adjust the range as per the number of pages
    # Construct the URL for the current page
    page_url = f"{base_url}{start_value}&type=1&type=1"  # Correcting the URL to have only one `&type=1`
    
    # Send the request to fetch the page content
    response = requests.get(page_url)
    
    # Ensure the page was fetched successfully
    if response.status_code != 200:
        print(f"Error: Unable to fetch the page {page_url}")
        continue
    
    # Parse the page content using BeautifulSoup
    page_data = BeautifulSoup(response.text, 'lxml')
    
    # Check if there are any tables in the page
    tables = page_data.find_all('table')
    
    # If there are no tables, skip this page
    if not tables:
        print(f"No tables found on page {page_url}")
        continue

    # If it's the first page (start_value == 0), use the second table (index 1)
    if start_value == 0:
        if len(tables) > 1:
            Individual_Test_Solutions_Table = tables[1]
        else:
            print(f"Second table not found on the first page {page_url}")
            continue
    else:
        # For subsequent pages, just access the first table (index 0)
        Individual_Test_Solutions_Table = tables[0]
    
    # Extract the rows from the table
    Test_per_page = Individual_Test_Solutions_Table.find_all('tr')[1:]

    # Loop through each test on the current page
    for test in Test_per_page:
        data_id = test['data-entity-id'] if test.get('data-entity-id') else "NONE"
        
        Test_Name = test.find('td', class_='custom__table-heading__title').text.strip() if test.find('td', class_='custom__table-heading__title') else "NONE"
        Test_Link = test.find('td', class_='custom__table-heading__title').find('a')['href'] if test.find('td', class_='custom__table-heading__title') else "NONE"
        print(Test_Name)
        tds = test.find_all('td', class_='custom__table-heading__general')
        
        Remote_Testing = "YES" if tds and tds[0].find('span', class_='catalogue__circle -yes') else "NO"
        Adaptive_Testing = "YES" if tds and len(tds) > 1 and tds[1].find('span', class_='catalogue__circle -yes') else "NO"
        
        Test_Types = []
        category_map = {
            'A': 'Ability & Aptitude',
            'B': 'BioData & Situational Judgement',
            'C': 'Competencies',
            'D': 'Development & 360',
            'E': 'Assessment Exercises',
            'K': 'Knowledge & Skills',
            'P': 'Personality & Behaviour',
            'S': 'Simulations'
        }
        
        SubTable = test.find('td', class_='custom__table-heading__general product-catalogue__keys')
        
        if SubTable:
            for letter, test_type in category_map.items():
                if letter in SubTable.text:
                    Test_Types.append(test_type)
                    
        response_test = requests.get('https://www.shl.com'+Test_Link).text
        pageData = BeautifulSoup(response_test, 'lxml')
        # Find all divs with the class 'product-catalogue-training-calendar__row typ'
        infos = pageData.find_all('div', class_='product-catalogue-training-calendar__row typ')
        
        # Initialize variables to store the information
        description = "NONE"
        job_levels = "NONE"
        languages = "NONE"
        assessment_length = "NONE"
        
        # Loop through each div element
        for info in infos:
            # Check the h4 text to determine the type of info
            h4_text = info.find('h4').text.strip() if info.find('h4') else "NONE"
            
            # Extract data based on h4 title
            if "Description" in h4_text:
                description = info.find('p').text.strip() if info.find('p') else "NONE"
            elif "Job levels" in h4_text:
                job_levels = info.find('p').text.strip() if info.find('p') else "NONE"
            elif "Languages" in h4_text:  # Assuming you are checking for Languages
                languages = info.find('p').text.strip() if info.find('p') else "NONE"
            elif "Assessment length" in h4_text:  # Assuming you are checking for Assessment Length
                assessment_length = info.find('p').text.strip() if info.find('p') else "NONE"
        
        # Now print the collected data
        # print(f"Description: {description}")
        
        test_data = {
            'Test_id': data_id,
            'Test_Name': Test_Name,
            'Test_Link': Test_Link,
            'Remote_Testing': Remote_Testing,
            'Adaptive_Testing': Adaptive_Testing,
            'Test_Types': Test_Types,
            'Description': description,
            'Job_Levels': job_levels,
            'Languages': languages,
            'Assessment_Length': assessment_length
        }
        
        all_tests_data.append(test_data)
```

```python
import csv

# Define the CSV fieldnames (these will be the column headers)
fieldnames = ['Test_id', 'Test_Name', 'Test_Link', 'Remote_Testing', 'Adaptive_Testing', 
              'Test_Types', 'Description', 'Job_Levels', 'Languages', 'Assessment_Length']

# Specify the file name where data will be saved
csv_file = 'test_data.csv'

# Open the CSV file in write mode
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    # Create a DictWriter object
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # Write the header (field names)
    writer.writeheader()
    
    # Write the rows (data for each test)
    for test in all_tests_data:
        writer.writerow(test)

print(f"Data has been written to {csv_file}")
```
